```python
# [과제] 자유주제로 프로그래밍을 수행하세요. 단 예외처리 3건 이상

ojall = {'과자' : 20 , '커피' : 30 , '김밥' : 5}
ojprice = {'과자' : 2000, '커피' : 1500, '김밥': 2500}
while 1:
    ojkey = list(ojall)
    print('='*120)
    print('''
                                           재고관리 프로그램
    ''')
    print('''                            항목                갯수           가격\n''')
    for i in range(len(ojkey)):
        print(f'''                          {ojkey[i]}        :      {ojall[ojkey[i]]}개       :   1개 {ojprice[ojkey[i]]}원''')
    print('\n')
    print('='*120)
    what = input("""무엇을 하시겠습니까?  
    
|  1 : 재고 추가  |  2 : 재고 제거  |  3 : 항목 추가  |  4 : 항목 제거  |  5 :  가격 변경  |
|  6 : 재고 검색  |  7 ; 재고 현황  |  0 : 프로그램 종료   |  \n """  )
    if what == '1':
        while 1:
            try:
                name = input("재고를 추가할 항목 이름을 입력해주세요.")
                ojall[name]
                while 1:
                    try:
                        mount = int(input("얼마나 추가 하시겠습니까   :   "))
                    except ValueError:
                        print('숫자를 입력해주세요')
                    else:
                        ojall[name] += mount
                        break
            except KeyError:
                print("없는 항목입니다. 다시 입력해주세요")
            else:
                print(f'{name} 항목에 {mount}개 추가되었습니다.\n')
                break
        
    if what == '2':
        while 1:
            try:
                name = input("재고를 제거할 항목 이름을 입력해주세요.")
                ojall[name]
                while 1:
                    try:
                        mount = int(input("얼마나 제거 하시겠습니까   :   "))
                    except ValueError:
                        print('숫자를 입력해주세요')
                    else:
                        ojall[name] -= mount
                        break
            except KeyError:
                print("없는 항목입니다. 다시 입력해주세요")
            else:
                print(f'{name} 항목에 {mount}개 제거되었습니다.\n')
                break
            
    if what == '3':
        name = input("추가할 항목 이름을 입력해주세요.")
        while 1:
            try:
                mount = int(input("현재 갯수를 입력해주세요   :   "))
            except ValueError:
                print('숫자를 입력해주세요')
            else:
                ojall[name] = mount
                break
        while 1:
            try:
                price = int(input("1개의 가격을 입력해주세요   :   "))
            except ValueError:
                print('숫자를 입력해주세요')
            else:
                ojprice[name] = price
                break
        print(f'{name} 항목 {mount}개,1개에 {price}원 추가되었습니다.\n')
            
    elif what == '4':
        while 1:
            try:
                name = input("제거할 항목 이름을 입력해주세요.   :  ")
                ojall[name]
            except KeyError:
                print('없는 항목입니다. 다시 입력해주세요')
            else:
                del ojall[name]
                del ojprice[name]
                print(f'{name} 항목 제거되었습니다.\n')
                break
                
    elif what == '5':
        while 1:
            try:
                name = input("가격을 변경할 항목 이름을 입력해주세요.")
                ojall[name]
                while 1:
                    try:
                        price = int(input("얼마로 변경 하시겠습니까   :   "))
                    except ValueError:
                        print('숫자를 입력해주세요')
                    else:
                        ojprice[name] = price
                        break
            except KeyError:
                print("없는 항목입니다. 다시 입력해주세요")
            else:
                print(f'{name} 항목의 가격이 {price}원으로 변경되었습니다.\n')
                break
                
    elif what == '6':
        while 1:
            try:
                name = input("검색 할 항목 이름을 입력해주세요.  :  ")
                ojall[name]
            except KeyError:
                print('없는 항목입니다. 다시 입력해주세요')
            else:
                mount = ojall[name]
                price = ojprice[name]
                print(f'{name}의 남은 재고 개수는 {mount}개, 1개의 가격은 {price}원, 재고의 총 가격은 {mount*price}원입니다.\n')
                break
                
    elif what == '7':
        print('재고 가격 순으로 정렬합니다.')
        list_ojmount = []
        list_ojmount = []
        temp1 = {}
        temp2 = {}
        for i,v in ojall.items():
            temp1[v*ojprice[i]] = i
            temp2[i] = v*ojprice[i]
            list_ojmount.append(v*ojprice[i])
        list_ojmount.sort()
        list_ojmount.reverse()
        for i in range(len(list_ojmount)):
            print('''%d.:\t%s %d개 %d원'''%(i+1,temp1[list_ojmount[i]],ojall[temp1[list_ojmount[i]]],temp2[temp1[list_ojmount[i]]]))
        print('재고의 총 가격은 %d원 입니다.\n' %sum(temp2.values()))
            
            
    elif what == '0':
        print('프로그램을 종료합니다.')
        break
        
    else:
        print("잘못 입력하셨습니다. 0 ~ 7 중 하나로 입력해주세요.\n")
```
