```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

train_df = pd.read_excel('./traindata.xlsx')
train_df
```

```python
test_df = pd.read_excel('./testdata.xlsx')
test_df
```

```python
def prepre(df):
    df.drop('custID',axis=1,inplace = True)
    return df
```

```python
prepre(train_df)
prepre(test_df)
train_df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split



X_train = train_df.drop('target',axis=1)
y_train = train_df['target']
X_test = test_df.drop('target',axis=1)
y_test = test_df['target']
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
def scores(y_test,pred=None,pred_proba=None):
    con_mat = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test,pred_proba)
    print(f'오차행렬 : \n{con_mat}\n정확도 : {acc:.4f}, 정밀도 : {pre:.4f}, 재현율 : {rec:.4f}, f1 점수 : {f1:.4f}, roc_auc : {roc_auc:.4f}')
```

```python
def model_score(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    pred_proba = model.predict_proba(X_test)[:,1]
    con_mat = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test,pred_proba)
    print(f'모델 : {model} \n 오차행렬 : \n{con_mat}\n정확도 : {acc:.4f}, 정밀도 : {pre:.4f}, 재현율 : {rec:.4f}, f1 점수 : {f1:.4f}, roc_auc : {roc_auc:.4f}')
```

```python
dt = DecisionTreeClassifier(random_state=123)
rf = RandomForestClassifier(random_state=123)
xgb = XGBClassifier(random_state=123)
lgbm = LGBMClassifier(random_state=123)
lr = LogisticRegression(random_state=123)

X_train,X_test,y_train,y_test = train_test_split(train_df.drop('target',axis=1),test_df['target'],test_size=0.2,random_state=123)

models = [dt,rf,lr,xgb,lgbm]
for i in models:
    model_score(i,X_train,y_train,X_test,y_test)
```

```python
rf = RandomForestClassifier(random_state=123)
rf.fit(X_train,y_train)
pred = rf.predict(X_test)
pred_proba = rf.predict_proba(X_test)[:,1]
scores(y_test,pred,pred_proba)
```

```python
X_test = test_df.drop('target',axis=1)
y_test = test_df['target']

pred1 = rf.predict(X_test)
pred_proba1 = rf.predict_proba(X_test)[:,1]
scores(y_test,pred1,pred_proba1)
```
