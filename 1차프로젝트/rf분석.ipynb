```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

train_df = pd.read_excel('./traindata.xlsx')
train_df
```

```python
test_df = pd.read_excel('./testdata.xlsx')
test_df
```

```python
from sklearn.preprocessing import LabelEncoder
def prepre(df):
    le = LabelEncoder()
    df['성별'] = le.fit_transform(df['성별'])
    df['연령대'] = le.fit_transform(df['연령대'])
    df['점포코드'] = le.fit_transform(df['점포코드'])
    df.drop('custID',axis=1,inplace = True)
    return df
```

```python
prepre(train_df)
prepre(test_df)
train_df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
X_train = train_df.drop('target',axis=1)
y_train = train_df['target']
X_test = test_df.drop('target',axis=1)
y_test = test_df['target']
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
def scores(y_test,pred=None,pred_proba=None):
    con_mat = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test,pred_proba)
    print(f'오차행렬 : \n{con_mat}\n정확도 : {acc:.4f}, 정밀도 : {pre:.4f}, 재현율 : {rec:.4f}, f1 점수 : {f1:.4f}, roc_auc : {roc_auc:.4f}')
```

```python
def model_score(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    pred = model.predict(X_test)
    pred_proba = model.predict_proba(X_test)[:,1]
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test,pred_proba)
    print(f'모델 : {model} \n 정확도 : {acc:.4f}, 정밀도 : {pre:.4f}, 재현율 : {rec:.4f}, f1 점수 : {f1:.4f}, roc_auc : {roc_auc:.4f}')
```

```python
rf = RandomForestClassifier()

rf.fit(X_train,y_train)
pred_rf = rf.predict(X_test)
pred_proba_rf = rf.predict_proba(X_test)[:,1]

scores(y_test,pred_rf,pred_proba_rf)
```

```python
from lightgbm import plot_importance
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
import seaborn as sns

font_path = '../m8_머신러닝응용/dataset/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family = font_name)

imfo_df = pd.DataFrame(X_train.columns.tolist(),columns=['name'])
imfo_df['importance']=rf.feature_importances_
s_df = imfo_df.sort_values(by='importance',ascending=False)
fig , ax = plt.subplots(figsize = (20,20))

sns.barplot(x=s_df.importance,y=s_df.name)
```

```python
from sklearn.model_selection import GridSearchCV
params ={
    'n_estimators':[10,50,100,200],
    'max_depth':[5,10,15,20,None],
    'min_samples_leaf':[8,12,18],
    'min_samples_split':[2,8,16,20]
}

rf = RandomForestClassifier(random_state=123)
# rf로 객체 생성 후 GridSearchCV 수행
grid_clf_f1 = GridSearchCV(rf,param_grid= params,cv=5,refit=True,scoring='f1',verbose=True)
grid_clf_f1.fit(X_train,y_train)
scores_df = pd.DataFrame(grid_clf_f1.cv_results_)

print('최적 하이퍼 파라미터',grid_clf_f1.best_params_)
print('f1 : ',grid_clf_f1.best_score_)
grid_pred_f1 = grid_clf_f1.best_estimator_.predict(X_test)
grid_pred_proba_f1 = grid_clf_f1.best_estimator_.predict_proba(X_test)[:,1]
scores(y_test,grid_pred_f1,grid_pred_proba_f1)
```
