```python
class Jego:
    def __init__(self):
        self.ojall = {'과자' : 20 , '커피' : 30 , '김밥' : 5}
        self.ojprice = {'과자' : 2500, '커피' : 2000, '김밥': 3000}
        self.ojkey = list(self.ojall)
        
    def jegoshow(self):
        self.ojkey = list(self.ojall)
        print('='*120)
        print('''
                                                        재고관리 프로그램
        ''')
        print('''                            항목                갯수           가격\n''')
        for i in range(len(self.ojkey)):
            print(f'''                          {self.ojkey[i]}        :      {self.ojall[self.ojkey[i]]}개       :   1개 {self.ojprice[self.ojkey[i]]}원''')
        print('\n')
        print('='*120)
        
    def jegoadd(self):
        while 1:
            name = input("재고를 추가할 항목 이름을 입력해주세요. 이전으로 돌아가시려면 e를 입력해주세요      :      ")
            if name in self.ojall:
                while 1:
                    try:
                        mount = int(input("얼마나 추가 하시겠습니까   :   "))
                    except:
                        print("잘못 입력하셨습니다. 숫자를 입력해주세요")
                    else:
                        while 1:
                            b = input(f'{name} 항목에 {mount}개 추가가 맞습니까?  y  or  n    :')
                            if b == 'y':
                                self.ojall[name] += mount
                                print(f'{name} 항목에 {mount}개 추가되었습니다.\n')
                                break
                            elif b == 'n':
                                print('취소되었습니다. 이전 화면으로 돌아갑니다.\n')
                                break
                            else:
                                print('y 나 n 로 입력해주세요.\n')
                        break
                break
            elif name == 'e':
                print('이전 화면으로 돌아갑니다.\n')
                break
            else:
                print("없는 항목입니다. 다시 입력해주세요.\n")
            
    def jegosub(self):
        while 1:
            name = input("재고를 제거할 항목 이름을 입력해주세요. 이전으로 돌아가시려면 e를 입력해주세요    :")
            if name in self.ojall:
                while 1:
                    try:
                        mount = int(input("얼마나 제거 하시겠습니까   :   "))  
                    except:
                        print("숫자를 입력해주세요")
                    else:
                        while 1:
                            b = input(f'{name} 항목에 {mount}개 제거가 맞습니까? y  or   n    : ')
                            if b == 'y':
                                self.ojall[name] -= mount
                                print(f'{name} 항목에 {mount}개 제거되었습니다.\n')
                                break
                            elif b == 'n':
                                print('취소되었습니다. 이전 화면으로 돌아갑니다.\n')
                                break
                            else:
                                print('y 나 n 로 입력해주세요.\n')
                        break
                break
            elif name == 'e':
                print('이전 화면으로 돌아갑니다.\n')
                break
            else:
                print("없는 항목입니다. 다시 입력해주세요.\n")
            
    def jegonew(self):
        name = input("추가할 항목 이름을 입력해주세요. 이전으로 돌아가시려면 e를 입력해주세요   : ")
        if name == 'e':
            print('이전 화면으로 돌아갑니다.\n')
        else:
            while 1:
                a = input(f'{name}이(가) 맞습니까? y or n : ')
                if a == 'y':
                    while 1:
                        try:
                            mount = int(input("현재 갯수를 입력해주세요 : "))
                        except:
                            print("숫자를 입력해주세요")
                        else:
                            break
                    while 1:
                        try:
                            price = int(input("1개의 가격을 입력해주세요 : "))
                        except:
                            print("숫자를 입력해주세요")
                        else:
                            break
                    while 1:
                        b = input(f'{name} 항목에 {mount}개, {price}원 맞습니까?   y  or  n   :  ')
                        if b == 'y':
                            self.ojall[name] = mount
                            self.ojprice[name] = price
                            print(f'{name} 항목 {mount}개,1개에 {price}원 추가되었습니다.\n')
                            break
                        elif b == 'n':
                            print('취소되었습니다. 이전 화면으로 돌아갑니다.\n')
                            break
                        else:
                            print('y 나 n 로 입력해주세요.\n')
                    break
                elif a == 'n':
                    print('취소되었습니다. 이전 화면으로 돌아갑니다.\n')
                    break
                else:
                    print('y 나 n 로 입력해주세요.\n')
      
    def jegodel(self):
        while 1:
            name = input("제거할 항목 이름을 입력해주세요. 이전으로 돌아가시려면 e를 입력해주세요   :  ")
            if name in self.ojall:
                while 1:
                    b = input(f'{name} 항목을 정말로 제거하시겠습니까? y or  n   : ')
                    if b == 'y':
                        del self.ojall[name]
                        del self.ojprice[name]
                        print(f'{name} 항목 제거되었습니다.\n')
                        break
                    elif b == 'n':
                        print('취소되었습니다. 이전 화면으로 돌아갑니다.\n')
                        break
                    else:
                        print('y 나 n 로 입력해주세요.\n')
                break
            elif name == 'e':
                print('이전 화면으로 돌아갑니다.\n')
                break
            else:
                print("없는 항목입니다. 다시 입력해주세요.\n")
       
    def jegosearch(self):
        while 1:
            name = input("검색 할 항목 이름을 입력해주세요. 이전으로 돌아가시려면 e를 입력해주세요   :  ")
            if name in self.ojall:
                mount = self.ojall[name]
                price = self.ojprice[name]
                print(f'{name}의 남은 재고 개수는 {mount}개, 1개의 가격은 {price}원, 재고의 총 가격은 {mount*price}원입니다.\n')
                break
            elif name == 'e':
                print('이전 화면으로 돌아갑니다.\n')
                break
            else:
                print("없는 항목입니다. 다시 입력해주세요.\n")
    
    def jegonow(self):
        print('재고 가격 순으로 정렬합니다.')
        list_ojmount = []
        list_ojmount = []
        temp1 = {}
        temp2 = {}
        for i,v in self.ojall.items():
            temp1[v*self.ojprice[i]] = i
            temp2[i] = v*self.ojprice[i]
            list_ojmount.append(v*self.ojprice[i])
        list_ojmount.sort()
        list_ojmount.reverse()
        for i in range(len(list_ojmount)):
            print('''%d.:\t%s %d개 %d원'''%(i+1,temp1[list_ojmount[i]],self.ojall[temp1[list_ojmount[i]]],temp2[temp1[list_ojmount[i]]]))
        print('재고의 총 가격은 %d원 입니다.\n' %sum(temp2.values()))
        
    def pricechange(self):
        while 1:
            name = input("가격을 변경할 항목 이름을 입력해주세요. 이전으로 돌아가시려면 e를 입력해주세요      :      ")
            if name in self.ojall:
                while 1:
                    try:
                        price = int(input("얼마로 변경 하시겠습니까   :   "))
                    except:
                        print("잘못 입력하셨습니다. 숫자를 입력해주세요")
                    else:
                        while 1:
                            b = input(f'{name} 항목의 가격을 {price}원으로 변경이 맞습니까?  y  or  n    :')
                            if b == 'y':
                                self.ojprice[name] = price
                                print(f'{name} 항목의 가격이 {price}원으로 변경되었습니다.\n')
                                break
                            elif b == 'n':
                                print('취소되었습니다. 이전 화면으로 돌아갑니다.\n')
                                break
                            else:
                                print('y 나 n 로 입력해주세요.\n')
                        break
                break
            elif name == 'e':
                print('이전 화면으로 돌아갑니다.\n')
                break
            else:
                print("없는 항목입니다. 다시 입력해주세요.\n")
    def start(self):
        while 1:
            self.jegoshow()
            what = input("""무엇을 하시겠습니까?  

|  1 : 재고 추가  |  2 : 재고 제거  |  3 : 항목 추가  |  4 : 항목 제거  |  5 :  가격 변경  |
|  6 : 재고 검색  |  7 ; 재고 현황  |  0 : 프로그램 종료   |  \n """  )
            if what == '1':
                self.jegoadd()
            elif what == '2':
                self.jegosub()
            elif what == '3':
                self.jegonew()
            elif what == '4':
                self.jegodel()
            elif what == '5':
                self.pricechange()
            elif what == '6':
                self.jegosearch()
            elif what == '7':
                self.jegonow()
            elif what == '0':
                print('프로그램을 종료합니다.')
                break
            else:
                print("잘못 입력하셨습니다. 0 ~ 7 중 하나로 입력해주세요.\n")
```
