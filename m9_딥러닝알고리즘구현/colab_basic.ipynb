#주요 Magic Command
pwd : 현재 위치  
ls : 현재 디렉토리 파일 조회  
whos : 현재 정의된 변수 확인  
reset : 현재 정의된 변수 삭제  
time : 실행시간을 알려줌

run : python file code 실행

matplotlib inline : matplotlib을 내부에서 실행할수 있도록 함
