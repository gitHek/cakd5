#### ÎèÖÎ¶ΩÎèôÏùºÎ∂ÑÌè¨
- ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥ÎùºÎäî Í≤ÉÏùÄ ÌôïÎ•†Î≥ÄÏàòÍ∞Ä Îã§Î•∏ ÌôïÎ•†Î≥ÄÏàòÏóê ÏòÅÌñ•ÏùÑ ÎÅºÏπòÏßÄ ÏïäÎäîÎã§Îäî ÏùòÎØ∏Ïù¥Îã§.
- ÌÜµÍ≥ÑÎ∂ÑÏÑùÏóêÏÑú Ï∑®Í∏âÌïòÎäî Îç∞Ïù¥ÌÑ∞ Ï§ë ÏÉÅÎãπÏàòÍ∞Ä ÎèÖÎ¶ΩÏù¥Î©¥ÏÑú Îã§Ï∞®ÏõêÏù∏ ÌôïÎ•†Î≥ÄÏàòÏùò Ïã§ÌòÑÍ∞íÏúºÎ°ú Í∞ÑÏ£ºÎê† Ïàò ÏûàÎã§.
- ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† Í∞ÅÍ∞Å ÎèôÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî Îã§Ï∞®Ïõê ÌôïÎ•†Î≥ÄÏàòÎ•º ÎèÖÎ¶ΩÎèôÏùºÎ∂ÑÌè¨Îùº ÌïúÎã§.

#### ÎèÖÎ¶ΩÏÑ±
- 2Í∞ú Ïù¥ÏÉÅÏùò ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏÑúÎ°ú ÏòÅÌñ•ÏùÑ ÎÅºÏπòÏßÄ ÏïäÏúºÎ©∞ Í¥ÄÍ≥ÑÍ∞Ä ÏóÜÏùåÏùÑ ÎÇòÌÉÄÎÇ¥Îäî Í∞úÎÖê
- 2Ï∞®Ïõê ÌôïÎ•†Î≥ÄÏàò (X,Y)Ïùò Í≤ΩÏö∞ Îã§ÏùåÍ≥º Í∞ôÏùÄ Í¥ÄÍ≥ÑÍ∞Ä ÏÑ±Î¶ΩÌï† Îïå X,YÎäî ÎèÖÎ¶ΩÏù¥ÎùºÍ≥† Ìï®.
$f_{X,Y}(x,y) = f_X(x)f_Y(y)$
- ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÎèÖÎ¶ΩÏùº Îïå Í≤∞Ìï©ÌôïÎ•†ÏùÄ Ï£ºÎ≥ÄÌôïÎ•†Ïùò Í≥±ÏúºÎ°ú ÏûëÏÑ±Ìï† Ïàò ÏûàÎã§.
- ÏùºÎ∞òÌôî : $f_{X_1,X_2,...X_n}(x_1,x_2,...,x_n) = f_{X_1}(x_1)f_{X_2}(x_2)...f_{X_n}(x_n)$

‚Äª Ìï®Ïàò fÍ∞Ä Ïù¥ÏÇ∞ÌòïÏù¥Î©¥ ÌôïÎ•†Ìï®Ïàò, Ïó∞ÏÜçÌòïÏù¥Î©¥ Î∞ÄÎèÑÌï®ÏàòÎ•º ÎÇòÌÉÄÎÉÑ

#### ÎèÖÎ¶ΩÏÑ±Í≥º Î¨¥ÏÉÅÍ¥ÄÏÑ±
- Í≥µÎ∂ÑÏÇ∞Ïù¥ÎÇò ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä 0Ïùº Îïå Î¨¥ÏÉÅÍ¥ÄÏù¥Îùº ÌïòÍ≥† Îëê ÌôïÎ•†Î≥ÄÏàò ÏÇ¨Ïù¥ÏóêÎäî ÏÉÅÍ¥ÄÏÑ±, Ï¶â ÏÑ†ÌòïÍ¥ÄÍ≥ÑÍ∞Ä ÏóÜÏùåÏùÑ ÎÇòÌÉÄÎÉÑ
- ÎèÖÎ¶ΩÏÑ±ÎèÑ Î¨¥ÏÉÅÍ¥ÄÏÑ±ÎèÑ Îëê ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏÑúÎ°ú Í¥ÄÍ≥ÑÍ∞Ä ÏóÜÎã§Îäî Í≤ÉÏùÑ ÎÇòÌÉÄÎÇ¥Îäî ÏÑ±ÏßàÏù¥Îã§.
    * Î¨¥ÏÉÅÍ¥ÄÏÑ±Î≥¥Îã§ ÎèÖÎ¶ΩÏÑ±Ïù¥ Îçî Í∞ïÌïú Í∞úÎÖê
    * 2Í∞úÏùò ÌôïÎ•†Î≥ÄÏàò XÏôÄ YÍ∞Ä ÎèÖÎ¶ΩÏùº Îïå XÏôÄ YÎäî Î¨¥ÏÉÅÍ¥ÄÏù¥ ÎêòÏßÄÎßå XÏôÄ YÍ∞Ä Î¨¥ÏÉÅÍ¥ÄÏùºÎïå XÏôÄ YÍ∞Ä Î∞òÎìúÏãú ÎèÖÎ¶ΩÏù∏Í≤ÉÏùÄ ÏïÑÎãò

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
linestyles = ['-', '--', ':', '-.']

def E(XY, g):
    x_set, y_set, f_XY = XY
    return np.sum([g(x_i, y_j) * f_XY(x_i, y_j)
                   for x_i in x_set for y_j in y_set])

def Cov(XY):
    x_set, y_set, f_XY = XY
    mean_X = E(XY, lambda x, y: x)
    mean_Y = E(XY, lambda x, y: y)
    return np.sum([(x_i-mean_X) * (y_j-mean_Y) * f_XY(x_i, y_j)
                    for x_i in x_set for y_j in y_set])

def f_X(x):
    return np.sum([f_XY(x, y_k) for y_k in y_set])

def f_Y(y):
    return np.sum([f_XY(x_k, y) for x_k in x_set])
```

```python
# Î∂àÍ≥µÏ†ïÌïú Ï£ºÏÇ¨ÏúÑ(ÍµêÏû¨ p130)

x_set = np.array([1, 2, 3, 4, 5, 6])
y_set = np.array([1, 2, 3, 4, 5, 6])

def f_XY(x, y):
    if x in x_set and y in y_set:
        return x * y / 441
    else:
        return 0
    
XY = [x_set, y_set, f_XY]
```

```python
# 2Ï∞®Ïõê ÌôïÎ•†Î≥ÄÏàò XÏôÄ YÎäî ÎèÖÎ¶Ω
# Í≥µÎ∂ÑÏÇ∞Ïù¥ 0Ïù¥ÎØÄÎ°ú Î¨¥ÏÉÅÍ¥ÄÏù∏ Í≤ÉÏùÑ Ïïå Ïàò ÏûàÏùå
Cov(XY)
```

```python
# Î¨¥ÏÉÅÍ¥ÄÏù∏ 2Í∞úÏùò ÌôïÎ•†Î≥ÄÏàò ÏÉùÍ∞ÅÌï¥Î≥¥Ïûê
x_set = np.array([0, 1])
y_set = np.array([-1, 0, 1])

def f_XY(x, y):
    if (x, y) in [(0, 0), (1, 1), (1, -1)]:
        return 1 / 3
    else:
        return 0
    
XY = [x_set, y_set, f_XY]
```

```python
# Í≥µÎ∂ÑÏÇ∞ÏùÄ 0Ïù¥ÎØÄÎ°ú ÌôïÎ•†Î≥ÄÏàò X,YÎäî Î¨¥ÏÉÅÍ¥Ä
Cov(XY)
```

```python
# ÎèÖÎ¶ΩÏÑ±Ïùò Ï†ïÏùòÎäî ùëìùëã,ùëå(ùë•,ùë¶)=ùëìùëã(ùë•)ùëìùëå(ùë¶)Í∞Ä ÏÑ±Î¶ΩÌï† ÎïåÏù∏Îç∞ Ïù¥Í≤ΩÏö∞ÏÑ±Î¶ΩÏó¨Î∂Ä ÌôïÏù∏
# Î¨¥ÏÉÅÍ¥ÄÏù¥Ïó¨ÎèÑ ÎèÖÎ¶ΩÏÑ±Ïù¥ ÏÑ±Î¶ΩÌïòÏßÄ ÏïäÎäî Í≤ΩÏö∞Í∞Ä ÏûàÎã§Îäî Í≤ÉÏùÑ ÌôïÏù∏
f_X(0)*f_Y(0),f_XY(0,0)
```

## Ìï©Ïùò Î∂ÑÌè¨
- ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† ÎèôÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÌôïÎ•†Î≥ÄÏàò $X_1,X_2,...X_nÏùò Ìï© \sum_{i=1}^n X_i = X_1 + X_2 + ... + X_n$ Ïù¥ Îî∞Î•¥Îäî ÌôïÎ•†Î∂ÑÌè¨Î•º ÏùòÎØ∏
- Ìï©Ïùò Î∂ÑÌè¨Î•º Ïù¥Ìï¥ÌïòÎ©¥ ÌôïÎ•†Î≥ÄÏàòÏùò Ìï©ÏùÑ ÎÇòÎàà ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨ÎèÑ ÏâΩÍ≤å Ïù¥Ìï¥Ìï† Ïàò ÏûàÏùå

```python
# Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Ìï©Ïùò Î∂ÑÌè¨
# ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ X ~ N(1,2), Y ~ N(2,3)Ïù¥ ÏûàÏùÑ Îïå ÌôïÎ•†Î≥ÄÏàò X + YÏùò Î∂ÑÌè¨Îäî N(3,5)Î•º Îî∞Î¶Ñ
rv1 = stats.norm(1, np.sqrt(2))
rv2 = stats.norm(2, np.sqrt(3))

sample_size = int(1e6)
X_sample = rv1.rvs(sample_size)
Y_sample = rv2.rvs(sample_size)

sum_sample = X_sample + Y_sample
print('ÌèâÍ∑†:', np.mean(X_sample),np.mean(Y_sample))
print('Î∂ÑÏÇ∞:', np.var(X_sample),np.var(Y_sample))
print('Ìï©Ïùò ÌèâÍ∑†, Î∂ÑÏÇ∞:', np.mean(sum_sample), np.var(sum_sample))
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

rv = stats.norm(3, np.sqrt(5))
xs = np.linspace(rv.isf(0.995), rv.isf(0.005), 100)

ax.hist(sum_sample, bins=100, density=True,
        alpha=0.5, label='N(1, 2) + N(2, 3)')
ax.plot(xs, rv.pdf(xs), label='N(3, 5)', color='gray')
ax.plot(xs, rv1.pdf(xs), label='N(1, 2)', ls='--', color='gray')
ax.plot(xs, rv2.pdf(xs), label='N(2, 3)', ls=':', color='gray' )

ax.legend()
ax.set_xlim(rv.isf(0.995), rv.isf(0.005))
plt.show()
```

Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò Ìï©Ïùò Î∂ÑÌè¨
- ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ X ~ Poi(3)Í≥º Y ~ Poi(4)Í∞Ä ÏûàÏùÑ Îïå ÌôïÎ•†Î≥ÄÏàò X + YÎèÑ Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Î•º Îî∞Î•¥Í≥† X + Y ~ Poi(7)Ïù¥ ÎêúÎã§.
- nÍ∞úÏùò Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò Ìï©Ïù¥ ÎêòÎçîÎùºÎèÑ Ïû¨ÏÑ±ÏÑ±Ïóê ÏùòÌïòÏó¨ Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§.

```python
rv1 = stats.poisson(3)
rv2 = stats.poisson(4)

sample_size = int(1e6)
X_sample = rv1.rvs(sample_size)
Y_sample = rv2.rvs(sample_size)
sum_sample = X_sample + Y_sample
print('ÌèâÍ∑†:', np.mean(X_sample), np.mean(Y_sample))
print('Î∂ÑÏÇ∞:', np.var(X_sample), np.var(Y_sample))
print('Ìï©Ïùò ÌèâÍ∑†, Î∂ÑÏÇ∞:', np.mean(sum_sample), np.var(sum_sample))
```

```python
# ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ nÍ∞úÏùò ÌôïÎ•†Î≥ÄÏàòÍ∞Ä Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïóê ÎåÄÌïòÏó¨ Í∑∏ Ìï©ÎèÑ Ïû¨ÏÉùÏÑ±Ïóê ÏùòÌï¥ Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò ÌòïÌÉúÎ•º Ïú†ÏßÄÌïúÎã§.
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(10, 6))
rv = stats.poisson(7)
xs = np.arange(20)
hist, _ = np.histogram(sum_sample, bins=20,
                       range=(0, 20), density=True)

plt.bar(xs, hist, alpha=0.5, label='Poi(3) + Poi(4)')
plt.plot(xs, rv.pmf(xs), label='Poi(7)',  color='gray')
plt.plot(xs, rv1.pmf(xs), label='Poi(3)', ls='--', color='gray')
plt.plot(xs, rv2.pmf(xs), label='Poi(4)', ls=':',  color='gray')

plt.legend()
plt.xlim(-0.5, 20)
plt.xticks(np.arange(20))
plt.show()
```

Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Ïùò Ìï©Ïùò Î∂ÑÌè¨
- Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨ÏóêÎäî Ïû¨ÏÉùÏÑ±Ïù¥ ÏóÜÎã§.
- Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Ïùò Ìï©ÏùÄ Ïù¥Ìï≠Î∂ÑÌè¨Í∞Ä ÎêòÎäî ÏÑ±ÏßàÏù¥ ÏûàÎã§.

```python
p = 0.3
rv = stats.bernoulli(p)

sample_size = int(1e6)
X_sample = rv.rvs((10, sample_size))

print('ÌëúÎ≥∏Ï∂îÏ∂ú:',X_sample.shape,'\n','Ï≤´Î≤àÏß∏ Ïó¥:', X_sample[:,0],'\n')
sum_sample = np.sum(X_sample, axis=0)
print('ÏÑ±Í≥µ ÌöüÏàò:', sum_sample,'\n')
print('ÏÑ±Í≥µ ÌöüÏàòÏùò ÌèâÍ∑†, Î∂ÑÏÇ∞:', np.mean(sum_sample), np.var(sum_sample))
print('ÏÑ±Í≥µ ÌöüÏàòÏùò ÌèâÍ∑†, Î∂ÑÏÇ∞:', np.mean(sum_sample), np.var(sum_sample,ddof=1))
```

```python
# normed=True Î•º ÎÑ£Ïñ¥Ï£ºÏñ¥ Ï†ïÍ∑úÌôîÎ•º ÌïòÎ©¥ Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ Í∞úÏàò Ï§ëÏóê Ìï¥Îãπ Íµ¨Í∞ÑÏùò Í∞í Í∞úÏàòÍ∞Ä Ï∞®ÏßÄÌïòÎäî ÎπÑÏú®Î°ú Î≥ÄÌôò
plt.figure(figsize=(10, 6))

rv = stats.binom(10, p)
xs = np.arange(10)
hist, _ = np.histogram(sum_sample, bins=10,
                       range=(0, 10), normed=True)
plt.bar(xs, hist, alpha=0.5, label='sum of 10*Bern(0.3)')
plt.plot(xs, rv.pmf(xs), label='Bin(10, 0.3)')
plt.legend()
plt.xlim(-0.5, 10)
plt.xticks(np.arange(10))
plt.show()
```

ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨
- ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨ÎûÄ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† ÎèôÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÌôïÎ•†Î≥ÄÏàò $X_1,X_2,...X_n$Ïùò ÌëúÎ≥∏ÌèâÍ∑†Ïù¥ Îî∞Î•¥Îäî Î∂ÑÌè¨
- Ïù¥ Î∂ÑÌè¨Îäî Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ÎÇò ÎåÄÏàòÏùò Î≤ïÏπô Îì± Ï∂îÏ∏°ÌÜµÍ≥ÑÏóêÏÑú Ï§ëÏöîÌïú ÏÑ±ÏßàÏùÑ Îã¥Í≥† ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê Ïú†ÏùòÌï¥Ïïº Ìï®.  
$E(\bar{X}) = \mu$  
$V(\bar{X}) = {\sigma}^2 / n$

```python
# Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌëúÎ≥∏ÌèâÍ∑† Î∂ÑÌè¨

mean = 1
var = 2
rv = stats.norm(mean, np.sqrt(var))

n = 10
sample_size = int(1e6)
X_sample = rv.rvs((n, sample_size))
print('ÌëúÎ≥∏ ÏÇ¨Ïù¥Ï¶à 10, ÌëúÎ≥∏ Í∞úÏàò 100000:', X_sample.shape,'\n')
sample_mean = np.mean(X_sample, axis=0)
print('ÌëúÎ≥∏ÌèâÍ∑† Î∞±ÎßåÍ∞ú:', sample_mean, len(sample_mean),'\n')
print('ÌëúÎ≥∏ÌèâÍ∑† Î∞±ÎßåÍ∞úÏùò ÌèâÍ∑†, Î∂ÑÏÇ∞:',np.mean(sample_mean), np.var(sample_mean))
```

```python
plt.figure(figsize=(10, 6))

rv_true = stats.norm(mean, np.sqrt(var/n))
xs = np.linspace(rv_true.isf(0.999), rv_true.isf(0.001), 100)
plt.hist(sample_mean, bins=100, density=True,
        alpha=0.5, label='sample mean of 10*N(1, 2)')
plt.plot(xs, rv_true.pdf(xs), label='N(1, 0.2)', color='gray')

plt.legend()
plt.xlim(rv_true.isf(0.999), rv_true.isf(0.001))
plt.show()
```

```python
# [Í≥ºÏ†ú] Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò ÌëúÎ≥∏ÌèâÍ∑† Î∂ÑÌè¨Îäî Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Î•º Îî∞Î•¥ÏßÄ ÏïäÍ≥† Ï¢åÏö∞ ÎåÄÏπ≠Ïù∏ Ï†ïÍ∑úÎ∂ÑÌè¨Ïóê Í∑ºÏÇ¨Ìïú ÌòïÌÉúÍ∞Ä ÎêòÎäîÍ≤ÉÏùÑ ÏûÖÏ¶ù ÌïòÏÑ∏Ïöî.
l = 3
rv = stats.poisson(l)

n = 10
sample_size = int(1e6)
X_sample = rv.rvs((n, sample_size))
sample_mean = np.mean(X_sample, axis=0)

plt.figure(figsize=(10, 6))

rv = stats.norm(3,np.sqrt(0.3))
x_s = np.linspace(0,6,100)
plt.plot(x_s,rv.pdf(x_s),label = 'N(3,0.3)')
plt.hist(sample_mean, bins=50, density=True,
        alpha=0.5, label='sample mean of 10*Poi(3)')

plt.legend()
plt.xlim(0, 6)
plt.show()
```

#### Ìï©Ïùò Î∂ÑÌè¨ÏôÄ ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨
- Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ÎÇò Ï†ïÍ∑úÎ∂ÑÌè¨ Í∞ôÏùÄ Ïû¨ÏÉùÏÑ±Ïù¥ ÏûàÎäî ÏùºÎ∂Ä ÌôïÎ•†Î∂ÑÌè¨Î•º Ï†úÏô∏ÌïòÎ©¥ Ìï©Ïùò Î∂ÑÌè¨Í∞Ä Ïñ¥Îñ§ ÌòïÌÉúÎ•º Î≥¥Ïù¥ÎäîÏßÄ ÏùºÎ∞òÏ†ÅÏúºÎ°ú Ïïå Ïàò ÏóÜÎã§.
- ÌëúÎ≥∏ÌèâÍµ∞Ïùò Î∂ÑÌè¨Í∞Ä ÎêòÎ©¥ Ïû¨ÏÑ±ÏÑ±ÏùÑ ÏßÄÎãå Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ Ï°∞Ï∞®ÎèÑ Í∑∏ ÌòïÌÉúÎ•º Ïú†ÏßÄÌï† Ïàò ÏóÜÎã§.
- ÏõêÎûò Î∂ÑÌè¨Í∞Ä Î¨¥ÏóáÏù¥ÏóàÎì†ÏßÄ Í∞ÑÏóê ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨Îäî Ï†ïÍ∑úÎ∂ÑÌè¨Ïóê Í∑ºÏÇ¨ÌïúÎã§


#### Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨
- ÌôïÎ•†Î≥ÄÏàò $X_1,X_2,...,X_n$Ïù¥ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† Í∏∞ÎåìÍ∞íÏù¥ $\mu, Î∂ÑÏÇ∞Ïù¥ {\sigma}^2$Ïù∏ ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•º Îïå nÏù¥ Ïª§ÏßêÏóê Îî∞Îùº ÌëúÎ≥∏ÌèâÍ∑† $\bar{X}$Ïùò Î∂ÑÌè¨Îäî Ï†ïÍ∑úÎ∂ÑÌè¨ $N(\mu,{\sigma}^2/n)$Ïóê Í∞ÄÍπåÏõåÏßÑÎã§.

```python
# [Í≥ºÏ†ú] Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨Î•º ÏûÖÏ¶ùÌïòÏÑ∏Ïöî.
# ÏÑ±Í≥µÌôïÎ•†Ïù¥ 0.3Ïù∏ Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨
p = 0.3
rv = stats.bernoulli(p)
# 10000Í∞úÏî© 10000Í∞ú ÏÉòÌîå ÌèâÍ∑† Íµ¨ÌïòÍ∏∞
n = 10000
sample_size = 10000
X_sample = rv.rvs((n, sample_size))
sample_mean = np.mean(X_sample, axis=0)
# Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Ïùò ÌèâÍ∑† = p , Î∂ÑÏÇ∞ = p(1-p)
rv_true = stats.norm(p, np.sqrt(p*(1-p)/n))
xs = np.linspace(rv_true.isf(0.999), rv_true.isf(0.001), 100)
plt.figure(figsize=(10, 6))

plt.hist(sample_mean, bins=50, density=True,
        alpha=0.5, label='sample mean of 10000*Bern(0.3)')
plt.plot(xs, rv_true.pdf(xs), label='N(0.3, 0.21/10000)', color='gray')

plt.legend()
plt.xlim(rv_true.isf(0.999), rv_true.isf(0.001))
plt.show()
```

#### ÎåÄÏàòÏùò Î≤ïÏπô
- ÌôïÎ•†Î≥ÄÏàò$X_1,X_2,...,X_n$Ïù¥ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† ÌèâÍ∑†Ïù¥ $\mu, Î∂ÑÏÇ∞Ïù¥ {\sigma}^2$Ïùº Í≤É Í∞ôÏùÄ ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•º Îïå, nÏù¥ Ïª§ÏßêÏóê Îî∞Îùº ÌëúÎ≥∏ÌèâÍ∑†ÏùÄ $\mu$Ïóê ÏàòÎ†¥ÌïúÎã§.

```python
# [Í≥ºÏ†ú] sample_sizeÏùò Î≥ÄÌôîÏóê Îî∞Îùº ÌëúÎ≥∏ÌèâÍ∑†Ïù¥ Î™®ÌèâÍ∑†Ïóê ÏàòÎ†¥Ìï®ÏùÑ ÏûÖÏ¶ùÌïòÏÑ∏Ïöî.
m = 3
v = 0.5
rv = stats.norm(3,np.sqrt(0.5))
x_s = np.linspace(100,100000,1000).astype(int)
p1 = np.zeros(len(x_s))
p2 = np.zeros(len(x_s))
p3 = np.zeros(len(x_s))
p4 = np.zeros(len(x_s))
for i in range(len(x_s)):
    p1[i] = rv.rvs(x_s[i],random_state = 1).mean()
    p2[i] = rv.rvs(x_s[i],random_state = 2).mean()
    p3[i] = rv.rvs(x_s[i],random_state = 3).mean()
    p4[i] = rv.rvs(x_s[i],random_state = 4).mean()
    
plt.figure(figsize=(10, 8))
plt.plot(x_s,p1)
plt.plot(x_s,p2)
plt.plot(x_s,p3)
plt.plot(x_s,p4)
plt.hlines(3,0,100000,color = 'black')
```
