```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

train_df = pd.read_excel('./traindata.xlsx')
train_df
```

```python
test_df = pd.read_excel('./testdata.xlsx')
test_df
```

```python
from sklearn.preprocessing import LabelEncoder
def prepre(df):
    le = LabelEncoder()
    df['성별'] = le.fit_transform(df['성별'])
    df['연령대'] = le.fit_transform(df['연령대'])
    df['점포코드'] = le.fit_transform(df['점포코드'])
    df.drop('custID',axis=1,inplace = True)
    return df
```

```python
prepre(train_df)
prepre(test_df)
train_df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LogisticRegression
X_train = train_df.drop('target',axis=1)
X_train60 = train_df[list60]
X_train40 = train_df[list40]
X_train20 = train_df[list20]
y_train = train_df['target']
X_test = test_df.drop('target',axis=1)
X_test60 = test_df[list60]
X_test40 = test_df[list40]
X_test20 = test_df[list20]
y_test = test_df['target']
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
def scores(y_test,pred=None,pred_proba=None):
    con_mat = confusion_matrix(y_test,pred)
    acc = accuracy_score(y_test,pred)
    pre = precision_score(y_test,pred)
    rec = recall_score(y_test,pred)
    f1 = f1_score(y_test,pred)
    roc_auc = roc_auc_score(y_test,pred_proba)
    print(f'오차행렬 : \n{con_mat}\n정확도 : {acc:.4f}, 정밀도 : {pre:.4f}, 재현율 : {rec:.4f}, f1 점수 : {f1:.4f}, roc_auc : {roc_auc:.4f}')
```

```python
rf60 = DecisionTreeClassifier(max_depth=40,min_samples_leaf=8,min_samples_split=2,random_state=156)

rf60.fit(X_train60,y_train)
pred_rf = rf60.predict(X_test60)
pred_proba_rf = rf60.predict_proba(X_test60)[:,1]

scores(y_test,pred_rf,pred_proba_rf)
```

```python
rf40 = DecisionTreeClassifier(max_depth=40,min_samples_leaf=8,min_samples_split=2,random_state=156)

rf40.fit(X_train40,y_train)
pred_rf = rf40.predict(X_test40)
pred_proba_rf = rf40.predict_proba(X_test40)[:,1]

scores(y_test,pred_rf,pred_proba_rf)
```

```python
rf20 = DecisionTreeClassifier(max_depth=40,min_samples_leaf=8,min_samples_split=2,random_state=156)
rf20.fit(X_train20,y_train)
pred_rf = rf20.predict(X_test20)
pred_proba_rf = rf20.predict_proba(X_test20)[:,1]

scores(y_test,pred_rf,pred_proba_rf)
```
